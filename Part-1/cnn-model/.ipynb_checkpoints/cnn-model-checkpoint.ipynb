```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

# Veri seti yolları
data_dir = "/kaggle/input/fruits/fruits-360_dataset_100x100/fruits-360"
train_dir = f'{data_dir}/Training'
test_dir = f'{data_dir}/Test'

# Model yolu
model_path = "/kaggle/input/cnn/keras/default/1/fruits_cnn_model.keras"

# Modeli yükle
model = tf.keras.models.load_model(model_path)

# Veri hazırlığı için ImageDataGenerator kullanımı
test_datagen = ImageDataGenerator(rescale=1./255)

# Test verilerini yükle
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Tahminleri yap
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# Metrikleri hesapla
accuracy = accuracy_score(y_true, y_pred_classes)
precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Sonuçları yazdır
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")


```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Veri seti yolları
data_dir = "/kaggle/input/fruits/fruits-360_dataset_100x100/fruits-360"
test_dir = f'{data_dir}/Test'

# Model yolu
model_path = "/kaggle/input/cnn/keras/default/1/fruits_cnn_model.keras"

# Modeli yükle
model = tf.keras.models.load_model(model_path)

# Veri hazırlığı için ImageDataGenerator kullanımı
test_datagen = ImageDataGenerator(rescale=1./255)

# Test verilerini yükle
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Tahminleri yap
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Confusion Matrix'i hesapla
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# 141 sınıfı 20 eşit parçaya böl
num_classes = len(class_labels)
split_size = 20
splits = [class_labels[i:i + split_size] for i in range(0, num_classes, split_size)]

# Parçaların confusion matrix'lerini oluştur
for i, split in enumerate(splits):
    split_indices = [test_generator.class_indices[label] for label in split]
    split_matrix = conf_matrix[np.ix_(split_indices, split_indices)]
    
    # Heatmap ile görselleştir
    plt.figure(figsize=(10, 8))
    sns.heatmap(split_matrix, annot=True, fmt='d', xticklabels=split, yticklabels=split, cmap='Blues')
    plt.title(f'Confusion Matrix for Split {i + 1}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```
