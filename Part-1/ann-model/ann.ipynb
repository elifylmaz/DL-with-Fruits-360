```python
import numpy as np
import glob
import os

# 1. Zip dosyasını açma
zip_file_name = "/content/fruits-360.zip"  # ZIP dosyanızın yolu
extract_path = "/content/fruits-360_dataset"      # Çıkartılacak hedef klasör

import zipfile
if not os.path.exists(extract_path):  # Daha önce açılmadıysa
    with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"{zip_file_name} dosyası açıldı.")

# 2. Veri yolu tanımları
base_path = os.path.join(extract_path, "fruits-360_dataset_100x100/fruits-360/")
dim = 100  # Resim boyutu
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

# Parametreler
num_classes = 141
train_dir = "/content/fruits-360_dataset/fruits-360/Training"
test_dir = "/content/fruits-360_dataset/fruits-360/Test"

# Geliştirilmiş ANN Modeli
def build_ann(input_shape, num_classes):
    model = Sequential([
        Dense(1024, activation='relu', input_shape=(input_shape,)),  # PCA'dan gelen boyuta uygun giriş
        BatchNormalization(),
        Dropout(0.5),

        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),

        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),

        Dense(128, activation='relu'),
        Dropout(0.4),

        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# ImageDataGenerator ile veri artırma ve yeniden ölçeklendirme
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True
)
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Veri akışını oluşturma
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(100, 100),
    batch_size=64,
    class_mode='sparse'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=64,
    class_mode='sparse'
)

# PCA ve scaler işlemleri
scaler = StandardScaler()
pca = PCA(n_components=0.95)  # %95 varyansı koruyacak şekilde bileşen sayısını belirle

# Eğitim verisi
X_train, y_train = [], []
for X_batch, y_batch in train_generator:
    X_flat = X_batch.reshape(X_batch.shape[0], -1)
    if len(X_train) == 0:
        X_scaled = scaler.fit_transform(X_flat)
        X_pca = pca.fit_transform(X_scaled)
    else:
        X_scaled = scaler.transform(X_flat)
        X_pca = pca.transform(X_scaled)
    X_train.append(X_pca)
    y_train.append(y_batch)
    if len(X_train) * train_generator.batch_size >= train_generator.samples:
        break

X_train = np.vstack(X_train)
y_train = np.hstack(y_train)

# Test verisi
X_test, y_test = [], []
for X_batch, y_batch in test_generator:
    X_flat = X_batch.reshape(X_batch.shape[0], -1)
    X_scaled = scaler.transform(X_flat)
    X_pca = pca.transform(X_scaled)
    X_test.append(X_pca)
    y_test.append(y_batch)
    if len(X_test) * test_generator.batch_size >= test_generator.samples:
        break

X_test = np.vstack(X_test)
y_test = np.hstack(y_test)

# PCA ile elde edilen bileşen sayısı
def get_pca_input_shape(pca):
    return pca.n_components_

input_shape = get_pca_input_shape(pca)

# Modeli oluştur ve özetle
model = build_ann(input_shape=input_shape, num_classes=num_classes)
model.summary()

# Callback'ler: EarlyStopping ve ReduceLROnPlateau
early_stopping = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)

# Modeli PCA ile dönüştürülmüş verilerle eğitme
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=64,
    callbacks=[early_stopping, reduce_lr]
)

# Eğitim ve doğrulama sonuçlarını görselleştirme
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Doğruluk grafiği
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Kayıp grafiği
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
# Modeli test verisi üzerinde değerlendirme
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")

# Sınıf raporunu yazdırma (Precision, Recall, F1-Score)
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)

report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()

# Precision, Recall, F1 için tabloyu yazdırma
print("Detailed Classification Report")
print(df_report)

# Confusion matrix'i görselleştirme
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=test_generator.class_indices.keys(), yticklabels=test_generator.class_indices.keys())
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Sınıfları 20'lik gruplara ayırma
num_classes = 141  # Toplam 141 sınıf
group_size = 20
num_groups = num_classes // group_size  # İlk 140 sınıfı 20'lik gruplara ayırıyoruz

# Son grup için kalan sınıfı ekleyeceğiz
extra_classes = num_classes % group_size  # Kalan sınıf sayısı (1 sınıf)

# Her grup için confusion matrix hesaplama ve görselleştirme
for i in range(num_groups):
    start_class = i * group_size
    end_class = (i + 1) * group_size

    # Bu grup içindeki y_test ve y_pred değerlerini filtreleyelim
    y_test_group = np.array([y if start_class <= y < end_class else -1 for y in y_test])
    y_pred_group = np.array([y if start_class <= y < end_class else -1 for y in y_pred])

    # -1 olanları çıkaralım (geçersiz sınıflar)
    valid_indices = (y_test_group != -1) & (y_pred_group != -1)
    y_test_group = y_test_group[valid_indices]
    y_pred_group = y_pred_group[valid_indices]

    # Confusion matrix hesaplama
    cm_group = confusion_matrix(y_test_group, y_pred_group, labels=range(start_class, end_class))

    # Görselleştirme
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm_group, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(start_class, end_class),
                yticklabels=range(start_class, end_class))
    plt.title(f"Confusion Matrix - Classes {start_class+1} to {end_class}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# Son grup için (Kalan sınıf) Confusion Matrix
if extra_classes > 0:
    start_class = num_groups * group_size
    end_class = num_classes

    # Bu grup içindeki y_test ve y_pred değerlerini filtreleyelim
    y_test_group = np.array([y if start_class <= y < end_class else -1 for y in y_test])
    y_pred_group = np.array([y if start_class <= y < end_class else -1 for y in y_pred])

    # -1 olanları çıkaralım (geçersiz sınıflar)
    valid_indices = (y_test_group != -1) & (y_pred_group != -1)
    y_test_group = y_test_group[valid_indices]
    y_pred_group = y_pred_group[valid_indices]

    # Confusion matrix hesaplama
    cm_group = confusion_matrix(y_test_group, y_pred_group, labels=range(start_class, end_class))

    # Görselleştirme
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm_group, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(start_class, end_class),
                yticklabels=range(start_class, end_class))
    plt.title(f"Confusion Matrix - Classes {start_class+1} to {end_class}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```
